```python
import ollama
def evaluate(s_ys:str,tot_ys:str,cot_ys:str):
    s_eval_text=str(s_ys)
    tot_eval_text=str(tot_ys)
    cot_eval_text=str(cot_ys)
    prompt=[{"role": "user", 
             "content": f'''You are a writing teacher. 
             The text below are the writing assignments from three students c,tot and cot.
             Read the passage written by the three students carefully. Plans, or ideas, any text besides the passage itself will not be evaluated.
             Your job is to determine the best, the second, and the last of those writing works interms of their logic and coherency.
             You should return with a ranking that clearly state which is the best which is the second and which is the last. AND DO NOT GENERATE ANYTING ELSE!!!!!! 
             The text is provided below:
             
             -----------------------------------
             The work by studnet s:
             {s_eval_text}
             -----------------------------------
             The work by student tot:
             {tot_eval_text}
             -----------------------------------
             The work by student cot:
             {cot_eval_text}'''}]
    res=ollama.chat(model='llama3',messages=prompt,)
    txt=str(res['message']["content"])
    return txt
```

```python
def coherency(ys:str):
    eval_text=str(ys)
    prompt=[{"role": "user", 
             "content": f'''You are a teacher, your assignment is to generate a coherency score for the following text, which is a writing sample by a student.
             The coherency score goes from 0 to 10 with 10 being the most coherent. You should generate only a number for the coherency score and no explaination is required.
             The text is provided below:
             {eval_text}'''}]
    res=ollama.chat(model='llama3',messages=prompt,)
    txt=str(res['message']["content"])
    return txt
```
